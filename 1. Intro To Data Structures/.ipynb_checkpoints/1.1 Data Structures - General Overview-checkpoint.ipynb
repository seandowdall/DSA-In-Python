# Data Structures - General Overview

## What is a data structure?
A data structure models an object so that it can be stored and organised easily by the computer.

## Data structures can be:

**Immutable**: Meaning it DOES NOT CHANGE after creation.

**Mutable**: Meaning it facilitates operations to be performed on contents. These operations can be updates or queries performed on the content of the structure.

## Examples

##### Mutable Data Structures

```python
# Lists

my_list = [1, 2, 3] 

# you can change the elements of the list

my_list[0] = 4 #Â here we change the first index of the list to 4.

#Dictionaries

my_dict = {'a': 1, 'b': 2}

# you can add remove or modify key-value-pairs in the dict e.g.

my_dict['c'] = 3  # which adds a new key-value pair {'c': 3}
```

##### Immutable Data Structures

```python
# Tuples

my_tuple = (1, 2, 3) # once created you cannot change the elements of a tuple without throwing an error

# Strings 

my_string = "hello"
```

## Why would we not always use mutable data structures?

As mutable data structures require time and effort to model, space could also be a factor. This is an area we will look into further down the line.

It is also important to be aware that decisions on what data structure you choose, will have a big effect on the progress of the project in the future.

## Classification of Data Structures

## There are two types of Data Structure: Linear & Non Linear

### 1. Linear Data Structure

- Relates to how the info is stored in the data structure.
- Elements are arranged one after the otheror sequentially, reflecting the order that they were inputted.

##### Examples of Linear Data Structures

- Arrays

- Lists

- Queues

- Stacks


Some languages will also demand that similar types of data are stored in the same structure, for example:
- Integer Lists

- String Arrays

Other languages will allow for MIXED Arrays
